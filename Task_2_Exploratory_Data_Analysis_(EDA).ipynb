# 2.Create histograms and boxplots for numeric features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Titanic-Dataset.csv')
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived', 'Pclass']
sns.set(style="whitegrid")

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col].dropna(), kde=True, bins=30, color='skyblue')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

#  3.Use correlation matrix for feature relationships

```python
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

#  4.Identify patterns, trends, or anomalies in the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Titanic-Dataset.csv')
sns.set(style="whitegrid")
# -----------------------------patterns--------------------------------------------------------------------
# Survival by Gender
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sex', hue='Survived', palette='Set2')
plt.title('Survival Count by Gender')
plt.show()

# Survival by Passenger Class
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set1')
plt.title('Survival Count by Passenger Class')
plt.show()

# --------------------------------anomalies----------------------------------------------------------------

# Detect zero or negative fares
print("\nPassengers with Fare = 0:")
print(df[df['Fare'] == 0])

# Outliers in Fare using boxplot
sns.boxplot(data=df, x='Fare', color='lightblue')
plt.title('Fare Outliers')
plt.show()

# Outliers in Age
sns.boxplot(data=df, x='Age', color='salmon')
plt.title('Age Outliers')
plt.show()

# Unusual family sizes
df['FamilySize'] = df['SibSp'] + df['Parch']
print("\nPassengers with unusually(FamilySize > 6):")
print(df[df['FamilySize'] > 6])
```
